```python
import pytrends
from pytrends.request import TrendReq
import pandas as pd
import time
import datetime
from datetime import datetime, date, time

timeframe = 'today 1-m'
key_words_general = ['pandemic', 'lockdown', 'shutdown', 'virus', 'toilet paper']
key_words_germany = ['Schulschließung', 'Quarantäne', 'Grenzen', 'Ladenschließung', 'Ausgangssperre']
locations_europe = ['DE', 'UK', 'FR', 'ES', 'IT']
locations_world = []
locations_germany = ['DE-BE', 'DE-BW', 'DE-SH', 'DE-ST', 'DE-SN', 'DE-SL', 'DE-BY', 'DE-BB', 'DE-HB', 'DE-HH', 'DE-HE', 'DE-MV', 'DE-NI', 'DE-NW', 'DE-RP', 'DE-TH']
```

```python
pytrend = TrendReq()
trends_germany = []
for loc in locations_germany:
    print(loc)
    pytrend.build_payload(kw_list=key_words_germany, timeframe=timeframe, geo=loc)
    trends_germany.append(pytrend.interest_over_time())
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(4, 4, sharex='all', sharey='all', figsize=(15,15))
for i, loc in enumerate(locations_germany):
    xi = i%4
    yi = int(i//4)
    trends_germany[i].plot(ax=ax[xi,yi])
    ax[xi,yi].set_title(loc)
plt.show()
```
